## No.1

```python
# Masukan syntax jawaban di sini
import pandas as pd

dict = {'no':[1, 2, 3, 4, 5],
        'nama_siswa': ["Adi", "Budi", "Carli", "Dadan", "Endang"],
        'umur':[15, 16, 14, None, 16],
        'email': ["adi@gmail.com", "budi@gmail.com", "carli@rakamin.com", "dadan@rakamin.com", "endang@gmail.com"],
        'score_matematika' : [80, 90, 100, 95, 97],
        'score_ekonomi' : [90, 95, 100, 86, None],
        'score_olahraga' : [97, 95, 96, 100, 94]}

df = pd.DataFrame(dict)

df.head()
```

## No.2 

```python
# Masukan sytnax jawaban di sini
data_sorted = df.sort_values(by = ['score_matematika'], ascending = False)
data_lowest_score = data_sorted.iloc[4]
data_lowest_score
```

## No.3

```python
# Masukan syntax jawaban di sini
import numpy as np
data_siswa_gmail = df[(df['email'].str.lower().str.contains('gmail')) & (df['score_ekonomi'] != None)]
data_siswa_gmail
```

```python
avg_umur = np.mean(data_siswa_gmail['umur'])
print('Rata - Rata Umur Data Siswa Gmail : ', avg_umur)
```

## No.4

```python
# Masukan syntax jawaban di sini
additional_dict = {'no':[6, 7],
        'nama_siswa': ["Fauzan", "Guntur"],
        'umur':[15, 16],
        'email': ["fauzan@gmail.com", "guntur@gmail.com"],
        'score_matematika' : [80, 90],
        'score_ekonomi' : [90, 95],
        'score_olahraga' : [97, 95]}

additional_data = pd.DataFrame(additional_dict)
new_data = df.append(additional_data, ignore_index = True)
median_umur = new_data['umur'].median()
median_umur
```

## No.5

```python
# Masukan syntax jawaban di sini
dict_address = {'nama': ["Adi", "Budi", "Dadan", "Endang", "Guntur", "Halim"],
                'alamat' : ["DKI Jakarta", "DKI Jakarta", "Banten", "Banten", "Jawa Barat", "Aceh"]}

data_address = pd.DataFrame(dict_address)
merge_data = new_data.merge(data_address, left_on = 'nama_siswa', right_on = 'nama')
avg_address = merge_data[['alamat', 'umur']].groupby(['alamat']).mean()
avg_address
```

```python
umur_min = avg_address.sort_values(by = ['umur'], ascending = True)
umur_min.iloc[0]
```

```python
# Masukan syntax jawaban di sini
data_airlines = pd.read_excel('data_airlines.xlsx')
data_airports = pd.read_excel('data_airports.xlsx')
data_flights = pd.read_excel('data_flights.xlsx')
```

```python
data_flights.groupby('MONTH').agg({'FLIGHT_NUMBER' : ['count'],
                             'CANCELLED' : ['sum'],
                             'ELAPSED_TIME' : ['mean', 'min', 'max']
                            }
                           )
```

## No.7

```python
# Masukan syntax jawaban di sini
filter_flights = data_flights[data_flights['ORIGIN_AIRPORT'].isin(['JFK', 'ECP'])]
data_merge = pd.merge(data_airports, filter_flights, left_on = 'IATA_CODE' , right_on = 'DESTINATION_AIRPORT')
destination_data = data_merge.groupby('AIRPORT').agg({'IATA_CODE' : ['count'],
                                                      'DISTANCE' : ['mean']})
destination_data
```

```python
data_code = destination_data['IATA_CODE']
code_sort = data_code.sort_values(by = ['count'], ascending = False)
code_sort.iloc[0]
```

```python
# Masukan syntax jawaban di sini
revenue = pd.read_csv('01 telecom_revenue.csv')
usage = pd.read_csv('02 telecom_usage.csv')

revenue.head()
```

```python
usage.head()
```

```python
df_1 = revenue[(revenue['MonthlyRevenue'] > 10) & (revenue['Occupation'].isin(['Professional', 'Student', 'Crafts']))]
df_1
```

```python
df_2 = usage[(usage['CustomerCareCalls'].astype(int).isin(usage['CustomerCareCalls'] % 2 == 0)) & (usage['UnansweredCalls'] > usage['BlockedCalls'])]
df_2
```

```python
df_merge = pd.merge(df_1, df_2, on = 'CustomerID')
df_filter = df_merge.groupby('Occupation').agg({'CustomerID' : ['nunique'],
                                                      'DroppedCalls' : ['mean']})
df_filter
```

```python
dropped = df_filter['CustomerID']
sort = dropped.sort_values(by = ['nunique'], ascending = False)
sort.iloc[0]
```
